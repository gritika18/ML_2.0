## Approach 2

I'm going to make a redundant ML model for the table of 10

You're going to need `xTrain`, `yTrain`, `xTest` and `yTest` from the `x` and `y` lists

````python
print(f'This is x: {x}')
print(f'This is y: {y}')
````

````python
xTrain = x[:-5]  #Training Data
yTrain = y[:-5]  #Training Labels

xTest = x[-5:] #Training Data
yTest = y[-5:] #Training Labels

print(f'''
Training Data:

xTrain : {xTrain}
yTrain : {yTrain}

Testing Data:

xTest : {xTest}
yTest : {yTest}
''')
````

````python
import tensorflow as tf         # tensorflow ia a very good framework to develop ML models 
from tensorflow import keras
````

````python
#perceptron model

model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])  #unit is the no of neurons and input_shape is the shape of the input to your neuron
# in this case, we are sending only pne input at a time so input_shape=[1]
````

````python
model.compile(optimizer='adam', loss='mae')   #optimizers --> sgd,rmsprop,adamax,adagrad;  loss --> mse,mae (this is regression)
````

````python
numepochs=10000
model.fit(x=xTrain, y=yTrain, validation_data=(xTest, yTest), epochs=numepochs)
# x->training data,  y->training labels, validation_data-> always takes a tuple
#loss->how well the model is performing on training data, validation_loss->how well the model is performing on tyesting data
#val_loss and loss can be analogous to cost function
#Output:
#1/1 [==============================] - 0s 71ms/step - loss: 0.2244 - val_loss: 0.3292
#<tensorflow.python.keras.callbacks.History at 0x7fcac9f01fd0>
````
